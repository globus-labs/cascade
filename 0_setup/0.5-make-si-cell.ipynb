# Make a Simulation Cell with Si, Si-vacancy

```python
from pathlib import Path
from ase.build import bulk
```

Create the cell

```python
n = 4 # number of times to repeat the cell
```

```python
atoms = bulk('Si') * [n,n,n]
```

```python
atoms.positions.shape
```

```python
Path('initial-geometries').mkdir(exist_ok=True)
atoms.write(f'initial-geometries/si-perfect-{n}x{n}x{n}.vasp', sort=True)
```

Create the cacancy

```python
del atoms[0] # create the vacancy
```

```python
atoms.positions.shape
```

```python
atoms.write(f'initial-geometries/si-vacancy-{n}x{n}x{n}.vasp', sort=True)
```

```python
atoms.get_volume()
```

```python
from ase.io import read
old = read('initial-geometries/si-vacancy-2x2x2.vasp')
```

```python
atoms.get_volume() / old.get_volume()
```
