```python
fig, ax = plt.subplots(figsize=(1.67,1.27))
temps = np.linspace(3000, 10000, 5)

# Example placeholder boltzmann function
def boltzmann(x, T):
    kB = 8.617333262145e-5  # eV/K
    p = np.exp(-x / (kB * T))
    return p / p.sum()

x = np.linspace(0, 1, 1000)

# Normalize for colormap mapping
norm = plt.Normalize(vmin=min(temps), vmax=max(temps))
cmap = cm.get_cmap('coolwarm')

for T in temps:
    color = cmap(norm(T))
    ax.plot(x, boltzmann(x, T), color=color)

ax.set_ylabel('P(x)', size=12)
ax.set_xlabel('E(x)', size=12)

# Add colorbar linked to ax
sm = cm.ScalarMappable(cmap=cmap, norm=norm)
sm.set_array([])
cbar = fig.colorbar(sm, ax=ax)

ax.set_xticklabels([])
ax.set_yticklabels([])
cbar.set_ticks([3000, 10000])
cbar.set_ticklabels(['Cold', 'Hot'], size=12) 
#ax.set_xticks([])
#ax.set_yticks([])
plt.savefig('boltzmann.png', 
            dpi=500, 
            bbox_inches='tight', 
            transparent=True
           )

plt.show()
```
